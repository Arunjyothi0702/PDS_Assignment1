```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data_clean/clean_StudentPerformance.csv")
df
```

```python
plt.figure(figsize=(6,6))
sns.histplot(df['writing score'], bins=10, kde=True)
plt.title('scores in Histogram')
plt.xlabel('writing score')
plt.ylabel('math score')
plt.show()
```

```python
plt.figure(figsize=(5, 2))
sns.scatterplot(x='writing score', y='reading score', data=df)
plt.title('Scatter Plot in writing score vs. reading score')
plt.xlabel('writing score')
plt.ylabel('reading score')
plt.show()
```

```python



plt.figure(figsize=(8, 6))
sns.violinplot(x='lunch', y='writing score', data=df)
plt.title('Distribution of Writing Score by Lunch')
plt.xlabel('Lunch')
plt.ylabel('Writing Score')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='race/ethnicity', y='math score', data=df)
plt.title('Average math Score vs. Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Math Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Countplot for Gender')
plt.xlabel('Gender')
plt.ylabel('parental level of education')
plt.show()
```
